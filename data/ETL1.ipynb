```python
from datetime import datetime

# Clean up the data
def clean_cell(index,value):
    if index in (0,1,2):
        return datetime.strptime(value,'%d/%m/%Y').date()
    if index in (3,4,5):
        ival = int(value.replace('.0',''))
        # Although a decimal is specified: It is never used.
        assert ival==float(value), value
        return ival
    if index in (7,8):
        if value.lower()=='yes': return True
        if value.lower()=='no' : return False
        raise ValueError(value)
    return value
data = [rawdata[0]] + [ [clean_cell(i,value) for i,value in enumerate(row)] for row in rawdata[1:] ]
len(data),data[1]
```

```python
import string

# How variable is each column?
def setSize(columnIndex):
    values = [ row[columnIndex] for row in data[1:]]
    return len(set(values))
[ (string.uppercase[i],header,setSize(i)) for i,header in enumerate(data[0]) ]
```

```python
from collections import Counter

def summary(columnIndex):
    values = [ row[columnIndex] for row in data[1:] ]
    values = Counter(values)
    return values.most_common(5)
# BIG FAT DATA SUMMARY
[ (string.uppercase[i],header,setSize(i),summary(i),'--------') for i,header in enumerate(data[0]) ]
```

```python
# Create a proper hash
datahash = [ { data[0][i]:value for i,value in enumerate(row) } for row in data[1:] ]
datahash[0]
```

```python
# Calculate a clean Content & Country per row
def get_continent(row):
    location = row['Location Of Investee'].lower()
    geo = row['Geography Of Beneficiaries'].lower()
    if location=='africa': return 'Africa'
    if geo=='asia'  : return 'Asia'
    if location=='india' : return 'Asia'
    if location=='uk' or location[:3]=='uk,'  : return 'Europe'
    if location=='canada': return 'North America'
    if geo=='europe': return 'Europe'
    if location=='' and geo=='': return '(unknown)'
    raise ValueError(location,geo)
# Test it works
print set([ get_continent(row) for row in datahash ])
for row in datahash:
    row['continent'] = get_continent(row)
```

```python
# Calculate a country for each row
# Results: Appear to be useless...
def get_country(row):
    location = row['Location Of Investee'].lower()
    geo = row['Geography Of Beneficiaries'].lower()
    if location=='uk': return 'UK'
    if location[:3]=='uk,': return 'UK'
    if location=='india': return 'India'
    if location=='canada': return 'Canada'
    if geo=='africa': return '(unknown)'
    if geo=='asia': return '(unknown)'
    if geo=='europe': return '(unknown)'
    if geo=='': return '(unknown)'
    raise ValueError(location,geo)
print set([ get_country(row) for row in datahash ])
for row in datahash:
    row['country'] = get_country(row)
```

```python
# Calculate a year for each row
for row in datahash:
    row['year'] = row['Investment Commitment Date'].year
```

```python
import json

# Construct a SanKey diagram
# Interesting fields:
#   year
#   Product Type
#   Purpose of Investment
#   continent
#   Sector
# --
# Output
# {"nodes":[  {"name":"Agricultural 'waste'"},... ],
#  "links":[  {"source":0,"target":1,"value":124.729}, ..]}
names = []
totals = {}
def link_nodes(col1,col2):
    for row in datahash:
        a = str(row[col1])
        b = str(row[col2])
        assert a,row
        assert b,row
        if a not in names:
            names.append(a)
        if b not in names:
            names.append(b)
        a_index = names.index(a)
        b_index = names.index(b)
        totals[(a_index,b_index)] = totals.get((a_index,b_index),0) + row['Cash Invested']
#link_nodes('year','Product Type')
link_nodes('Type Of Organisation Invested In','Product Type')
link_nodes('Product Type','Purpose Of Investment')
link_nodes('Purpose Of Investment','Sector')

sankey = { 
        'nodes': [{'name':x} for x in names], 
        'links': [{'source':a_index,'target':b_index,'value':amount} for ((a_index,b_index),amount) in totals.iteritems()]
        }
with open('etl_sankey.json','w') as f:
    json.dump(sankey,f)
```

```python
# Summarise year-on-year as a stacked bar
def get_2013_total(fieldname,fieldvalue):
    out = 0
    for row in datahash:
        if row['year']==2013 and row[fieldname]==fieldvalue:
            out += row['Cash Invested']
    return out
            
major = 'year'
minor = 'Sector'
all_major_values = sorted(list(set([ x[major] for x in datahash if x[major]!=2008])))
all_minor_values = sorted(list(set([ x[minor] for x in datahash])), key=lambda x:get_2013_total(minor,x), reverse=True)
def get_series(major_value):
    totals = {}
    for row in datahash:
        if not row[major]==major_value: continue
        minor_value = row[minor]
        totals[minor_value] = totals.get(minor_value,0) + row['Cash Invested']
    return { 
            'major': str(major_value), 
            'ages':[{'name':x,'value':totals.get(x,0)} for x in all_minor_values]
            }
year_on_year = {}
year_on_year['legend'] = all_minor_values
year_on_year['series'] = [ get_series(x) for x in all_major_values ]
with open('etl_yearonyear.json','w') as f:
    json.dump(year_on_year,f,indent=4)
```

```python
# Summarise a pair of pie charts
# ---
# (1) Total per sector
sector_total = {}
for row in datahash:
    sector = row['Sector']
    value = row['Cash Invested']
    sector_total[sector] = sector_total.get(sector,0) + value
pie1 = [{'name':key,'value':value} for (key,value) in sorted(sector_total.items(),key=lambda x:x[1])]

with open('etl_pie1.json','w') as f:
    json.dump(pie1,f)
```

```python
# Summarise a pair of pie charts
# ---
# (2) Secured vs Unsecured
unsecured_total = {}
for row in datahash:
    type = row['Product Type']
    unsecured = 'unsecured' in type.lower()
    value = row['Cash Invested']
    unsecured_total[type] = unsecured_total.get(type,0) + value
def sortorder(x):
    # Look closely for the hack here
    if x[0]=='Loans and facilities - Unsecured': return -2
    if x[0]=='Loans and facilities - Partially secured': return -1
    return x[1]
pie2 = [{'name':key,'value':value} for (key,value) in sorted(unsecured_total.items(),key=sortorder)]

with open('etl_pie2.json','w') as f:
    json.dump(pie2,f)
```

```python
# Summarise CoInvestment
for row in datahash:
    row['coinvestment'] = row['Total Value'] - row['Cash Invested']
coinvestment_data = {}
coinvestment_data['total'] = sum([x['coinvestment'] for x in datahash])
coinvestment_data['points'] = []
for row in datahash:
    if row['coinvestment']==0: continue
    if row['year']<2010: continue
    coinvestment_data['points'].append({
        'date' : row['Investment Commitment Date'].isoformat(),
        'coinvestment' : row['coinvestment'],
        'cash' : row['Cash Invested'],
        'origin' : row['Source Of Investment']
    })
coinvestment_data['points'] = sorted(coinvestment_data['points'],key=lambda x:x['date'])
with open('etl_bubblechart.json','w') as f:
    json.dump(coinvestment_data,f)
sorted([ (x['Investment Commitment Date'],x['coinvestment']) for x in datahash], key=lambda x:x[0])
```
